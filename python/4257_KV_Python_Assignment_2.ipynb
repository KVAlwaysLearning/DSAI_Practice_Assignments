# **Q-1**

The Infinite Loop Trap Write a function countdown_skip_threes(start). Use a while loop to count down from start to 0. Inside the loop, if the current number is a multiple of 3, skip printing it by using the continue keyword. Otherwise, print the number. Trick: Be very careful where you place your code that decreases the counter. If you place it after the continue statement, what will happen? Ensure your function avoids this trap.

# **Answer**

```python
def countdown_skip_threes(start):
  s=start
  while s>0:
    if s%3==0:
      s=s-1
      continue
    else:
      print(s)
      s=s-1
```

Inner vs. Outer Escape Write a function find_target_pair(list_a, list_b, target). Use nested for loops to find a number from list_a and a number from list_b that add up to the target. Print the pair as a tuple (e.g., (2, 5)). Trick: Once you find the first correct pair, you must stop searching entirely. Since a break statement only exits the innermost loop, you must use a clever condition or boolean variable to ensure the outer loop breaks as well.

# **Answer**

```python
def find_target_pair(list_a, list_b, target):
  f=0
  for i in list_a:
    for j in list_b:
      if i+j  == target:
        # Pair Found
        print(f"({i},{j})")
        f=1
        break

    if f==1:
      # Pair is already Found...Stopping
      break

```

```python
find_target_pair([1,2,3], [1,2,3], 5)
```

# **Q-10**

The Compromised System Write a function validate_passwords(password_list). Loop through the list of strings.

If a password is less than 8 characters, use continue to move to the next one.

If a password contains a space (” ”), print ”SYSTEM BREACH” and use break to stop checking the entire list immediately.

If a password contains no numbers, use pass (assume we will write a number-checker later, but for now, let it slide).

If the loop processes a password without triggering continue or break, print ”Valid password: [password]”.



# **Answer**

```python
def validate_passwords(password_list):

  for p in password_list:
    if len(p)<8:
      continue
    elif any(i==" " for i in p):
      print("SYSTEM BREACH")
      break
    elif not any(i in "0123456789" for i in p):
      pass
    else:
      print(f"Valid password: [{p}]")
```

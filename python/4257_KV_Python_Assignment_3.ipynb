# **Q-3**

3. The Reversal Trick

Given s = ”madam racecar”, write a single expression using slicing that extracts the word ”racecar” and reverses it.

# **Answer**

```python
s="madam racecar"
# 'racecar' is a Palindrome.

# This code ensures that only racecar is extracted even if there is something after the word
s[s.find('racecar'):s.find('racecar')+len('racecar')+1][::-1]
```

```python
s[s.find('racecar'):][::-1]
```

5. Chained Operations

Given s = ”I love Python programming!”, find the index of ”Python”, slice out the word, reverse it, and then replace the lowercase ’o’ with a zero ’0’. Write the chained expression.

# **Answer**

```python
s = "I love Python programming!"
s[s.find("Python"):s.find("Python")+len("Python")+1][::-1].replace("o","0")
```

```python
s.replace("Python",s[s.find("Python"):s.find("Python")+len("Python")+1][::-1].replace("o","0"))
```

7. The ”Sandwich” Slicer

Given phrase = ”apple-banana-cherry”, use the index() method to find the positions of the two hyphens. Then, use slicing and string addition to swap ”apple” and ”cherry” to output ”cherry-banana-apple”.

# **Answer**

```python
phrase = "apple-banana-cherry"
# Both approaches use the idea of extracting only a single word before and after hyphen '-' regardless what is beyond those words
# Slicing and String Addition
phrase[-(phrase[::-1].index("-")):len(phrase)-(phrase[::-1].index("-"))+len('cherry')+1]+phrase[phrase.index('-'):phrase.index('-')+len('banana')+2]+phrase[phrase.index("-")-len('apple'):phrase.index("-")]
```

```python
# Only Slicing and NO string Addition
phrase.replace("apple",phrase[-(phrase[::-1].index("-")):len(phrase)-(phrase[::-1].index("-"))+len('cherry')+1])[::-1].replace("cherry"[::-1],phrase[phrase.index("-")-len('apple'):phrase.index("-")][::-1],1)[::-1]
```

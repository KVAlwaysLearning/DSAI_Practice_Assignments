# **Q-1**

Ask the user to input a single decimal number. Using only floor division (//) and typecasting, extract the integer portion of the number. Multiply the string ">>" by that integer and print it, ensuring the output ends exactly with the string "[END]" instead of a newline.

# **Final Answer**

```python
l=[] # List for Storing user inputs

def user_input():
  print("Type 'STOP' to stop the program")
  c=0 # Counter
  m=[]
  # Pattern Creator
  def process_num(num):
      user_int=int(num//1)
      pattern=user_int*(">>")
      print(pattern, end="")
      print("\n","Count of Repeating units of '>>' : ",int(len(pattern)/len(">>")), sep="", end="") # Output ends on the same-line
      print("") # Adde for fetching next input on new line
      return None

  user_num=(input("Enter a number: "))
  l.append(user_num)
  m=(user_num.split(" "))

  if m[0].lower() != "stop" :

    decimal_count=user_num.count(".") #count decimals


    for i in user_num:

      if i in ("0123456789."): #check each character
        c=c+1


    if c==len(user_num) and user_num != "":

      if decimal_count == 1 :
          num=float(user_num)
          print("User entered the float/non-integer number :", num)
          process_num(num)

      elif decimal_count == 0 :
          num=int(user_num)
          print("User entered the Integer number :", num)
          process_num(num)

      else: # Handling multi-decimal absurd number strings (i.e. IP Addresses)
          print("It's an Incorrect number-string. Please, enter a valid number")


    else: # for non-number or mixed inputs
      print("Please enter only a decimal or an integer number")


    user_input()

  else: # Final STOP
      print("User chose not to enter further input")
      print("User inputs so far have been :",l[:-1])
      return None


```
